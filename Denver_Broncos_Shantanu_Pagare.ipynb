### Q5. How many owners have at least two different kinds of pets (e.g. a dog and a cat)?

```python
df_pet_owner.head()
```

```python
# For this we need to group the data by 'OwnerID' and 'Kind', and count the unique kinds for each owner:
df_owners_pets_counts = df_pet_owner.groupby(['OwnerID', 'Kind'])['PetID'].nunique().reset_index()
df_owners_pets_counts
```

```python
# Grouping the data again by 'OwnerID', and counting the number of different kinds of pets for each owner
df_owners_different_kinds = df_owners_pets_counts.groupby('OwnerID')['Kind'].count()
df_owners_different_kinds
```

```python
# Filtering out owners with at least two different kinds of pets:
df_owners_with_two_different_kinds = df_owners_different_kinds[df_owners_different_kinds >= 2]
df_owners_with_two_different_kinds
```

```python
num_owners_with_two_different_kinds = len(df_owners_with_two_different_kinds)
num_owners_with_two_different_kinds
```

```python
print("\033[1mNumber of owners with at least two different kinds of pets are:", num_owners_with_two_different_kinds)
```

```python
# Step 4: Print the result
print("\033[1mNumber of owners who have pets with matching first letters of their names is:", 
      df_num_owners_with_matching_letters)
```

### Q8. What percentage of cities have more male pets than female pets?

```python
# Here grouping the data by 'City' and 'Gender', and then counting the number of pets both male and female:
df_pets_by_city_gender = df_pet_owner.groupby(['City', 'Gender'])['PetID'].count().reset_index()
df_pets_by_city_gender
```

```python
# Then pivoting the data to see indicidual males and females in each city:
df_pivoted_pets_by_city = df_pets_by_city_gender.pivot(index='City', columns='Gender', values='PetID')
df_pivoted_pets_by_city
```

```python
# Here Filling Nan values with zeros, to avoid data errors: 
df_pivoted_pets_by_city = df_pivoted_pets_by_city.fillna(0)
```

```python
# Here counting the number of male pets that are greater than the count of female pets:
df_cities_with_more_male_pets = df_pivoted_pets_by_city[df_pivoted_pets_by_city['male'] > df_pivoted_pets_by_city['female']]
df_cities_with_more_male_pets
```

```python
# Calculating the percentage of cities where there are more male pets:
percentage_cities_more_male = (len(df_cities_with_more_male_pets) / len(df_pivoted_pets_by_city)) * 100
```

```python
# Result: 57.50%
print("\033[1mPercentage of cities with more male pets than female pets:", percentage_cities_more_male, "%")
```

### Q9. Which city's pet sample is made up of exactly 70% dogs? The answer is case sensitive, so please match the value for City exactly.

```python
# Filtering the DataFrame to include only rows where the 'Kind' column is 'Dog':
df_dogs = df_pet_owner[df_pet_owner['Kind'] == 'Dog']
df_dogs.head()
```

```python
# Grouping the filtered data by 'City', and counting the number of dogs for each city:
df_dogs_by_city = df_dogs.groupby('City')['PetID'].count().reset_index()
df_dogs_by_city
```

```python
# How many total pets are in each city, like including dog, cat, parrot, etc.
df_pets_by_city = df_pet_owner.groupby('City')['PetID'].count().reset_index()
df_pets_by_city
```

```python
# To check if there are any  repeating PetIDs, none found:
df_pet_owner['PetID'].value_counts()
```

```python
df_dogs_and_pets = pd.merge(df_dogs_by_city, df_pets_by_city, on='City', suffixes=['_Dogs', '_Pets'])
df_dogs_and_pets
```

```python
# Calculating the Percentage:
df_dogs_and_pets['Percentage of Dogs'] = (df_dogs_and_pets['PetID_Dogs'] / df_dogs_and_pets['PetID_Pets']) * 100
df_dogs_and_pets
```

```python
df_exactly_70 = df_dogs_and_pets[df_dogs_and_pets['Percentage of Dogs'] == 70]['City']
df_exactly_70
```

```python
print("\033[1mThe name of the city which has excatly 70% of the dogs is: ", df_exactly_70.values)
```
